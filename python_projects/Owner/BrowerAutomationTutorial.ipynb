```python
# Source:  https://www.youtube.com/watch?v=ApA7EVwSzg0
# Splinter doc: https://splinter.readthedocs.io/en/latest/finding.html
# Selenium list of options: 
#   https://peter.sh/experiments/chromium-command-line-switches/
#   https://chromium.googlesource.com/chromium/src/+/master/chrome/common/chrome_switches.cc



from splinter import Browser
from selenium import webdriver
from config import fb_username, fb_password, fb_url
from config import twitter_username, twitter_password, twitter_url
from config import profitly_challenge_username, profitly_challenge_password, profitly_challenge_url


# define path to chrome driver
executable_path = {'executable_path':r'C:\Users\Owner\python_projects\chromedriver.exe'}

# Set some default behaviors for our browser
options = webdriver.ChromeOptions()

# Make sure the window is maximized
options.add_argument("--start-maximized")

# Make sure that notification are off
options.add_argument("--disable-notifications")


## Twitter Section
# create a new browser object, by default it is firefox
twitter_browser = Browser("chrome", **executable_path, headless = False, options = options)



twitter_browser.visit(twitter_url)

# find my email element
twitter_username_box = twitter_browser.find_by_name("session[username_or_email]")

# fill out the element
twitter_username_box.fill(twitter_username)

# find password element
twitter_password = twitter_browser.find_by_name("session[password]").fill(twitter_password)


# Find submit button
#twitter_login_button = twitter_browser.find_by_tag("div").find_by_tag("span").find_by_text("Log in")
twitter_login_button = twitter_browser.find_by_text("Log in")


# Click the first element we find
twitter_login_button.first.click()

## Facebook Section
# create a new browser object, by default it is firefox
fb_browser = Browser("chrome", **executable_path, headless = False, options = options)

fb_browser.visit("https://www.facebook.com")

# find my email element
username_box = fb_browser.find_by_id("email")

# fill out the element
username_box.fill(fb_username)

# find password element
password_box = fb_browser.find_by_id("pass")
password_box.fill(fb_password)

# Find submit button
login_button = fb_browser.find_by_id("u_0_b")

# Click the first element we find
login_button.first.click()




```
